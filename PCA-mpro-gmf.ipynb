```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(style="whitegrid")
```

```python
PCA_full=pd.read_csv("C:\\Users\\Avell\\OneDrive\\COVID-19\\Paper\\Final\\PCA-new\\PCA-geral\\dimer_merged_PCA_proj.csv")
N3=PCA_full.loc[1:12416:1]
Apo=PCA_full.loc[12417:24832:1]
N1=PCA_full.loc[24833:35946:1]
N3d=PCA_full.loc[35947:43872:1]
Apod=PCA_full.loc[43873:52289:1]
N1d=PCA_full.loc[52290:60405:1]

#Noligand-6lu7
plot = sns.jointplot(x="PC_1", y="PC_2", data= Apo, color = "silver", space=0, kind="kde")


plot.ax_marg_x.set_xlim(-70, 150)
plot.ax_marg_y.set_ylim(-70, 50)


fig1 = plt.gcf()
plt.rcParams['axes.labelsize'] = 24
plt.rcParams['axes.titlesize'] = 24
plt.xlabel("PC1 (82.8%)")
plt.ylabel("PC2 (4.3%)")
plt.show()
plt.draw()
fig1.savefig('Apo-mono.pdf', dpi=600)
```
