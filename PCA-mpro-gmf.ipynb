```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(style="whitegrid")
```

```python
Apo=pd.read_csv("C:\\Users\\Avell\\OneDrive\\COVID-19\\Paper\\Final\\PCA-new\\Monomer\\PCA-separate\\483\\483_PCA_proj.csv")


#Noligand-6lu7
plot = sns.jointplot(x="PC_1", y="PC_2", data= Apo, color = "silver", kind="kde")


plot.ax_marg_x.set_xlim(-120, 150)
plot.ax_marg_y.set_ylim(-40, 60)


fig1 = plt.gcf()
plt.rcParams['axes.labelsize'] = 24
plt.rcParams['axes.titlesize'] = 24
plt.xlabel("PC1 (25.9%)")
plt.ylabel("PC2 (9.3%)")
plt.show()
plt.draw()
fig1.savefig('Apo-mono.pdf', dpi=600)
```
